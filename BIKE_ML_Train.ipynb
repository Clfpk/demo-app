```python
import pandas as pd
df=pd.read_csv("bike_buyers_clean.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.drop(["ID","Gender"],axis=1)
df.head()
```

```python
df.dtypes
```

```python
df["Income"]=df["Income"].astype("float")
df.dtypes
```

```python
cclum=[]
for i in df.columns:
    if df[i].dtypes=="object":
        cclum.append(i)
print(cclum)
from sklearn import preprocessing
le = preprocessing.LabelEncoder()   
df[cclum]=df[cclum].apply(le.fit_transform)
df.head()
```

```python
x=df.drop(["Purchased Bike"],axis=1)
y=df["Purchased Bike"]
x=x.to_numpy()
x.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import confusion_matrix,classification_report
names=["SVM","NB","KNN","DT","RF","ET"]
clfs=[SVC(),GaussianNB(),KNeighborsClassifier(),DecisionTreeClassifier(),RandomForestClassifier(),ExtraTreesClassifier()]
CLFF=[]
for name,clf in zip(names,clfs):
    print(name)
    clf.fit(X_train,y_train)
    y_pred=clf.predict(X_test)
    CLFF.append(clf)
    print(confusion_matrix(y_test,y_pred))
    print(classification_report(y_test,y_pred))
    print('-----------------------------------------------')
```

```python
import pickle
pickle.dump(CLFF,open("model_bike.pkl","wb"))
```
